```python
import folium
import numpy as np
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
from folium import plugins
from folium.plugins import HeatMap
from folium.plugins import Choropleth
```

```python
data1 = pd.read_csv("./counties-age.csv")
```

```python
data1
```

```python
def geocode_address(address):
  geolocator = Nominatim(user_agent="my_geocoder")
  try:
    location = geolocator.geocode(address)
    if location:
      return location.latitude, location.longitude
    else:
      return None
  except GeocoderTimedOut:
    return geocode_address(address)
```

```python
data1["Location"] = data1["NAMELSAD"] + ", " + data1["STUSPS"]
data1["Latitude"], data1["Longitude"] = zip(*data1["Location"].apply(geocode_address).dropna())
```

```python
center_lat = data1["Latitude"].mean()
center_lon = data1["Longitude"].mean()
```

```python
m = folium.Map(location=[center_lat, center_lon], zoom_start=6)
```

```python
heatmap_data = list(zip(data1["Latitude"], data1["Longitude"], data1["Population Age 65 and Older"]))
HeatMap(heatmap_data).add_to(m)
```
